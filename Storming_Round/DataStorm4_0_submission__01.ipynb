**Handling Missing Values**

```python
his_transaction_data.isnull().sum()
```

```python
his_transaction_data["item_price"].describe()
```

```python
def impute_mode(group):
    try:
        mode_value = group.mode()[0]
    except:
        return group.fillna(np.nan)
    return group.fillna(mode_value)


his_transaction_data["item_description"] = his_transaction_data[["item_description", "item_price"]].groupby('item_price')["item_description"].transform(impute_mode)
```

```python
his_transaction_data.head(50)
```

```python
his_transaction_data.isnull().sum()
```

```python
# dropping the rows with missing values
his_df = his_transaction_data.dropna()

# copying the dropped values in to a new dataframe
dropped_data = his_transaction_data[~his_transaction_data.index.isin(his_df.index)]
# his_transaction_data.shape
```

**Removing Duplicate Values**

```python
his_df.drop_duplicates(inplace=True)
```

# **Feature Extraction**

```python
his_df.info()
```

```python
# extracting date as year, month, day, week
# extracting time 
# to seperate columns
his_df[["date", "time"]] = his_df["transaction_date"].str.split('T',expand=True)
his_df.drop(columns="time",inplace=True)
```

# **Feature Encoding**

```python
his_df["item_description"].describe()
```

item_description have 37 uniqure values. The categories aren't ordinal. Therefore using OHE encode the item_descriptioption

---



```python
prev_features = list(his_df.columns)
his_df = pd.get_dummies(his_df, columns = ["item_description"])
# his_df.head()
```

```python
features =  [x for x in list(his_df.columns) if x not in  prev_features ]

for i in features:
  his_df[i] = his_df[i]*his_df["quantity_sold"]
# his_df.head(50)
```

```python
# multiplying item price with number of items
his_df["total_price"] = his_df["item_price"]*his_df["quantity_sold"]
df = his_df.groupby(["shop_id"]).sum()
df.columns
```

# **Model Applying**

```python
final_df= pd.merge(df,store_info,on="shop_id", how = 'left')
# final_df.head(50)

```

```python
final_df.drop(columns ="invoice_id",axis=1, inplace=True)
final_df.drop(columns ="item_price",axis=1, inplace=True)
df_to_predict = pd.merge(testing_data.drop("shop_profile",axis=1),final_df, on ="shop_id", how="left")
final_df.drop(columns ="shop_id",axis=1, inplace=True)
df_to_predict.drop(columns ="shop_id",axis=1, inplace=True)

# df_to_predict.head(50)

final_df.head(50)
```

```python
#final_df.fillna(np.nan)
# test_df = final_df[final_df["shop_profile"].isnull()]
train_df = final_df[~final_df["shop_profile"].isnull()]
# test_df.head(50)
# train_df.head(50)
```

```python
# splitting the train_df into two
# into train and test dataset

X = train_df.drop("shop_profile",axis =1)
Y = train_df["shop_profile"]

# X.head()
```

```python
# Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.2, shuffle= True,random_state= 42)
```

**`Training the model `**


```python
from sklearn.ensemble import RandomForestClassifier

random_forest_clf = RandomForestClassifier(n_estimators= 5, random_state=0)

# model training
random_forest_clf.fit(X_train, Y_train)
```

**Predicting values and evaluating accuracy**

```python
y_predict = random_forest_clf.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test,y_predict)

print(accuracy)
```

```python
# Evaluating accuracy using F1 score

from sklearn.metrics import f1_score

f1 = f1_score(Y_test, y_predict, average ='macro')
 
print(f1)
```

**Predicting the test data**

```python
# predicting the shop_profile
X_to_predict = df_to_predict.drop(columns="shop_profile",axis=1)
y_test_predict = random_forest_clf.predict(X_to_predict)
y_test_predict = pd.DataFrame(y_test_predict)
y_test_predict.head(50)
```
